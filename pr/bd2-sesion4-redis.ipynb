## Primeros pasos con Redis <a name="primerospasos"></a>



Con la operación `set` se inserta una clave (`key`) con el valor indicado (`value`). Si la clave ya existe con un valor previo, se sobrescribe, independientemente de su tipo.

Con la operación `get` obtenemos el valor de la clave indicada (`key`).

````python
redis_client.set('key', b'value') #set es crear y modificar. Si existe lo machaca
data: bytes | None = redis_client.get('key') #get es par coger el valor
print(type(data))
print(data) #si haces un print directamente, siempre va a poner b'valor. La b delante y el valor
````

## Tipos de datos fundamentales

#### Strings

````python
redis_client.set("nombre", "Juan") #también podemos meterle un string en vez de un byte (b)
redis_client.set("años", '15')

nombre: bytes | None = redis_client.get("nombre")
años: bytes | None = redis_client.get("años")
print(nombre)
print(años)
````

Por defecto, Redis devuelve respuestas de tipo `byte`. Para poder decodificarlas al formato correcto debemos usar el método `.decode()` (por defecto se utiliza el juego de caracteres UTF-8).

````python
print(nombre.decode() if nombre else None) #hacer el decode para sacar el dato sin el b y sin comillas
print(años.decode() if años else None)
````

#### Listas

Redis permite asociar listas a una determinada clave con el método `.lpush()`. Con el método `.lrange()` se obtienen elementos de una lista según el rango indicado (-1 indica hasta el último). Por ejemplo:

````python
#lpush; crear una lista a partir de una clave donde yo a esa clave le voy a meter todos esos elementos que voy a querer meter. SOLO una clave
# meto por el inicio de la lista.

redis_client.lpush("nums", 1)
redis_client.lpush("nums", 2)
redis_client.lpush("nums", 3)

numbers: list = redis_client.lrange("nums", 0, -1) # me recorre la clave y quiero que me devuelva la posición 0 y el -1 es hasta el final
print(numbers)

#rpush: es lo mismo, pero lo que hace es insertar por la derecha (por el final). En este ejemplo seria: 1,2,3.

#Como peculiaridad: permite repetidos y tiene un órden.
````

#### Conjuntos

También se pueden asociar un conjunto (set) a una clave con el método `.sadd()`. El método `.smembers()` devuelve todos los miembros del conjunto.

````python
# En conjuntos no hay una posición concreta como en el array. Otra diferencia es que no me permite repeticiones. Eso implica que cada vez que añado uno, va a comprobar si está
redis_client.sadd("fruits", "apple")
redis_client.sadd("fruits", "banana")
redis_client.sadd("fruits", "orange")

fruits: set = redis_client.smembers("fruits")
print(fruits)
````

Ten en cuenta que una lista (`lpush`) mantiene el orden y permite elementos repetidos, mientras que un conjunto (`sadd`) no mantiene orden y no permite  duplicados.

A fin de realizar transacciones debemos de usar los siguientes comandos:
* `.watch()`: permite monitorizar las claves sobre las cuales vamos a trabajar.
* `.multi()`: inicia el almacenado de comandos en una lista.
* `.execute()`: ejecuta todos los comandos almacenados en la lista.

Dos ejemplos de monitorización de claves serían: Evitar duplicidades de claves mientras se da de alta un usuario en un sistema, que no se duplique el
username cuando dos usuarios se estén dando de alta al mismo tiempo. Otro ejemplo, que dos usuarios vayan a incrementar un voto positivo
sobre un post y haya una condición de carrera.

````python
from redis import WatchError

with redis_client.pipeline() as pipe:
    while True:
        try:
            pipe.watch('key1', 'key2', 'key3') #mira los valores antes de cambiarlos,
            pipe.multi() #entre el multi y execute, es lo que quiero que se ejecute

            pipe.set('key1', 'valor1')
            pipe.set('key2', 'valor2')
            pipe.set('key3', 'valor3')

            pipe.execute() #para mirar si coinciden con lo visto en el watch, si no, volvería a pedir
            break

        except WatchError:
            continue
````

````python
import asyncio
import os
from asyncio import Semaphore, Task, create_task, gather
from collections.abc import Awaitable, Iterator

from duckdb import DuckDBPyRelation
from pandas import DataFrame, Index, notna
from redis import Redis
from redis.client import Pipeline


def run_in_new_thread[T](coro: Awaitable[T]) -> asyncio.Future[T]:
    """Run a coroutine in a separate thread with its own event loop.
    Returns an awaitable Future. This avoids blocking the main event loop
    when the coroutine contains blocking calls (e.g. DuckDB, Redis)."""
    def _run():
        loop: asyncio.AbstractEventLoop = asyncio.new_event_loop()
        asyncio.set_event_loop(loop)
        try:
            return loop.run_until_complete(coro)
        finally:
            loop.close()
    return asyncio.get_event_loop().run_in_executor(None, _run)


async def load_dataframe_redis(
    prefix: str, df: DuckDBPyRelation, redis_client: Redis
) -> list[list]:
    sem: Semaphore = Semaphore(int((os.cpu_count() or 2) * 1.5))

    async def execute_add_query(
        rows: DataFrame, sem: Semaphore,
    ) -> list:
        async with sem:
            pipe: Pipeline = redis_client.pipeline(transaction=False)
            for row_dict in rows.to_dict('records'):
                row_dict: dict = {col: val for col, val in row_dict.items() if notna(val)}
                id: str = row_dict["Id"]

                pipe.hset(f"{prefix}{id}", mapping=row_dict) #añade todos los posts con un prefijo, con un id, y un diccionaio con todos los campos
            return pipe.execute()

    def generate_chunks(df: DuckDBPyRelation, sem: Semaphore) -> Iterator[Task[list]]:
        while True:
            chunk: DataFrame = df.fetch_df_chunk(10)  # 10*2048
            if chunk.empty:
                break

            # Convert all datetime columns to string before processing
            datetime_cols: Index = chunk.select_dtypes(include=['datetime64']).columns
            chunk[datetime_cols] = chunk[datetime_cols].astype(str)
            yield create_task(execute_add_query(chunk, sem))

    # Generate all chunks and execute them concurrently. Wait for all to complete.
    all_chunks: Iterator[Task[list]] = generate_chunks(df, sem)
    return await gather(*all_chunks)
````

Definimos una función de ayuda que sirve para hacer eficientes los recorridos por todas las claves:

````python
from collections.abc import Generator

#funcion para iterar en la BD. Le pasas la BD, el patron que quiero bsuscar, y cuántos quiero buscar
def keys_iterator(redis_client: Redis, pattern: bytes, count: int = 10000) -> Generator[Iterator[bytes]]:
    cursor: int = 0
    while True:
        cursor, keys = redis_client.scan(cursor=cursor, match=pattern, count=count) #cursor para saber por donde me he quedado
        if keys:
            yield keys
        if cursor == 0:
            break
    return None
````

Primero, recorremos todos los posts. Esto es muy lento, como dijimos, pero es necesario para construir el índice la primera vez. Para cada post, obtenemos su lista de tags, y por cada tag añadimos el Id del post al conjunto de posts de ese tag (`poststags:{tag}`) con `sadd`.

````python
import re

from tqdm import tqdm

# Procesamos todos los posts uno a uno
posts: Iterator = redis_client.scan_iter(match=b'post:*')
for p_key in tqdm(posts, desc="Procesando posts..."):
  post_tags: bytes|None = redis_client.hget(p_key, 'Tags')
  if not post_tags:
      continue

  # Encontrar todos los tags separados por <xxx>
  pipe: Pipeline = redis_client.pipeline(transaction=False)
  for tag in re.findall(r'<(.*?)>', post_tags.decode()): #dame lo que está entre < >
    pipe.sadd(f"poststag:{tag}", p_key) #el valor es el p_key que ha encontrado con el scan
  pipe.execute()
````

### Ejemplo 2: Posts de cada tag, pero por Id de tag, no por nombre

A la hora de crear un post, si necesitamos buscar el `Id` de cada Tag dentro, podríamos hacer lo siguiente.

**NÓTESE DOS COSAS**:

1. Primero se crean las entradas `tagname:tag` para cada `tag`, que en su contenido guarda el `Id` del tag. Esto es un índice secundario, que nos permite encontrar el `Id` del tag a partir de su nombre.
2. Segundo, para evitar tantas búsquedas en la base de datos, se construye la función `get_tag_id(tag_name)` que toma el nombre del tag y devuelve su `Id` correspondiente. Esta función utiliza una optimización denominada "memoization", que consiste en guardar en memoria el resultado de una función para no tener que volver a calcularlo. Esto se consigue usando el decorador `@cache` de `functools`. De esta forma, si se llama a la función con el mismo argumento, se devuelve el resultado almacenado en memoria en lugar de volver a calcularlo. Esto es especialmente útil cuando la función es costosa de calcular o se llama muchas veces con los mismos argumentos. Esta optimización **sólo se puede usar si sabemos que el número de elementos cabe en memoria, y además si sabemos que no va a cambiar mientras estamos haciendo la computación**. En este caso sí, porque se refiere a los tags (unos miles). Si no, existe otra construcción `@lru_cache` que permite almacenar en memoria un número limitado de resultados, y va eliminando los más antiguos.

````python
import re
from functools import cache

#Añade al sistema nuevas entradas de este tipo:
# Construimos las claves tagname:nombre -> id a partir de las claves tag:id -> {TagName: nombre}

id_tags: Iterator = redis_client.scan_iter(match=b'tag:*') #escanea todo lo que en la bd empiece por tag:
pipe: Pipeline = redis_client.pipeline(transaction=False)
for t in id_tags:
  tag: bytes|None = redis_client.hget(t, 'TagName')
  if not tag:
      continue
  pipe.set(f"tagname:{tag.decode()}", t)
pipe.execute()


def calc_post_tags_id(redis_client: Redis) -> None:

  @cache #se define una caché para poder ir más rápido
  def get_tag_id(tag_name: str) -> str:
      """Devuelve el Id del tag dado su nombre. Función memoizada.
      Evita buscar en la BD cuando la búsqueda ya se ha realizado."""
      return redis_client.get(f"tagname:{tag_name}").decode()

  post_keys_it: Generator[Iterator[bytes]] = keys_iterator(redis_client, b'post:*')
  for batch in tqdm(post_keys_it, desc="Procesando 10000 posts..."):
    pipe: Pipeline = redis_client.pipeline(transaction=False)
    for p_key in batch:
      post_tags: bytes|None = redis_client.hget(p_key, 'Tags')
      if not post_tags:
          continue

      # Encontrar todos los tags separados por <xxx>
      for tag in re.findall(r'<(.*?)>', post_tags.decode()):
        db_tagid: str = get_tag_id(tag)
        _, tag_id = db_tagid.split(':')  # Id del tag
        pipe.lpush(f"poststagid:{tag_id}", p_key)
    pipe.execute()
````

### Ejemplo 4: Ordenar los posts por su fecha de creación

Podríamos usar un `sorted set` para generar un índice secundario que ordenara los `posts` en base a su fecha de creación (`CreationDate`). Para este ejemplo usamos `zadd` que añade uno o varios elementos a un conjunto ordenado y les asigna un valor numérico que determina su posición dentro del conjunto. Ten en cuenta que, `sadd` añade elementos a un conjunto sin orden, mientras que `zadd` los guarda ordenados según el valor numérico asignado a cada elemento.

````python
from datetime import datetime

post_keys_it: Generator[Iterator[bytes]] = keys_iterator(redis_client, b'post:*')
for batch in tqdm(post_keys_it, desc="Procesando 10000 posts..."):
  pipe: Pipeline = redis_client.pipeline(transaction=False)
  for p_key in batch:
    p_date_str: str = redis_client.hget(p_key,'CreationDate').decode()
    p_dt: float = datetime.strptime(p_date_str, '%Y-%m-%d %H:%M:%S.%f').timestamp()
    #print(p_dt)
    pipe.zadd("post_by_creationdate", {p_key:p_dt})
  pipe.execute()
````

## Bloom Filters como índice probabilístico

Un **Bloom filter** es una estructura de datos probabilística que permite comprobar si un elemento *posiblemente* pertenece a un conjunto o *definitivamente* no pertenece.

**Características principales:**
- Muy eficiente en memoria: usa un *array de bits*
- **Falsos positivos posibles**: puede decir que un elemento está cuando no lo está
- **Falsos negativos imposibles**: si dice que no está, con seguridad no está
- Útil como **pre-filtro** para evitar consultas costosas innecesarias

**Funcionamiento:**
1. Se inicializa un array de $m$ bits, todos a 0
2. Se definen $k$ funciones hash independientes
3. **Inserción**: se calculan las $k$ posiciones hash y se ponen esos bits a 1
4. **Consulta**: se calculan las $k$ posiciones; si **todos** están a 1, el elemento *posiblemente* existe; si **alguno** está a 0, **definitivamente no** existe

La probabilidad de falso positivo es: $p \approx (1 - e^{-kn/m})^k$

### Implementación usando operaciones de bits de Redis

Redis soporta comandos de bits sobre strings. En este ejemplo usaremos `BITFIELD` para agrupar varias lecturas/escrituras de bits en un único acceso a Redis.

(primero instalamos la librería mmh3 para hacer _hash_ rápido sobre cadenas de caracteres)

````python
import mmh3  # MurmurHash3, librería de hashing rápido: pip install mmh3
from bitarray import bitarray  # Para gestionar eficientemente array de bits en memoria


class RedisBloomFilter:
    """Bloom filter en memoria con array de bits eficiente.
    Permite construir el filtro rápidamente en Python y luego
    copiarlo a Redis en una sola operación.
    """

    def __init__(self, size: int = 100000, num_hashes: int = 7):
        """Inicializa el Bloom filter en memoria.
        Args:
            size: tamaño del array de bits (m)
            num_hashes: número de funciones hash (k)
        """
        self.size: int = size
        self.num_hashes: int = num_hashes
        self.bits: bitarray = bitarray(size, endian='big')  # Todos los bits a 0 inicialmente
        self.bits.setall(0)

    def _get_positions(self, item: str) -> list[int]:
        """Calcula las k posiciones de hash para un elemento."""
        return [abs(mmh3.hash(item, i)) % self.size
                for i in range(self.num_hashes)]

    def add(self, item: str) -> None:
        """Añade un elemento al Bloom filter en memoria."""
        positions: list[int] = self._get_positions(str(item))
        for pos in positions:
            self.bits[pos] = 1

    def exists(self, item: str) -> bool:
        """Comprueba si un elemento posiblemente está en el filtro (en memoria).
        True = posiblemente existe (puede ser falso positivo)
        False = definitivamente NO existe
        """
        positions: list[int] = self._get_positions(str(item))
        return all(self.bits[pos] for pos in positions)

    def to_redis(self, client: Redis, key: str) -> None:
        """Copia el array de bits a Redis en una sola operación."""
        expected_nbytes: int = (self.size + 7) // 8
        client.set(key, self.bits.tobytes()[:expected_nbytes])

    @staticmethod
    def from_redis(client: Redis, key: str, size: int, num_hashes: int = 7) -> 'RedisBloomFilter':
        """Carga un Bloom filter desde Redis para comprobar membresía."""
        bits_bytes = client.get(key)
        if bits_bytes is None:
            raise ValueError(f"Bloom filter key '{key}' not found in Redis")

        expected_nbytes: int = (size + 7) // 8
        if len(bits_bytes) < expected_nbytes:
            raise ValueError(
                f"Bloom filter key '{key}' has insufficient bytes: "
                f"{len(bits_bytes)} < {expected_nbytes}"
            )

        loaded_bits: bitarray = bitarray(endian='big')
        loaded_bits.frombytes(bits_bytes[:expected_nbytes])
        loaded_bits = loaded_bits[:size]

        bf = RedisBloomFilter(size=size, num_hashes=num_hashes)
        bf.bits = loaded_bits
        return bf


print("Clase RedisBloomFilter definida correctamente.")
````

### Ejemplo: Bloom filter `user_id:tag_name` y conjuntos `tagsusers:<tag>`

Construiremos simultáneamente:

1. Conjuntos `tagsusers:<tag_name> -> {user_id1, user_id2, ...}` con los usuarios que han hecho una pregunta con ese tag.
2. Un único Bloom filter `bf:user_tag` con claves `"user_id:tag_name"` para cada par usuario-tag encontrado.

El Bloom filter nos permitirá responder rápidamente: *¿el usuario X ha hecho alguna pregunta con el tag Y?*

````python
# Construir conjuntos tagsusers:<tag> y Bloom filter bf:user_tag simultáneamente

BF_USER_TAG_KEY: str = 'bf:user_tag'
BF_USER_TAG_HASHES: int = 7
BF_USER_TAG_SIZE: int = 1_000_000  # 1M bits para pares user:tag

# Limpiar estructuras previas (si existen)
for k in redis_client.scan_iter(match=b'tagsusers:*'):
    redis_client.delete(k)
redis_client.delete(BF_USER_TAG_KEY)
````

````python
# Bloom filter en memoria
bf_user_tag = RedisBloomFilter(size=BF_USER_TAG_SIZE, num_hashes=BF_USER_TAG_HASHES)

pares_insertados: int = 0
tags_procesados: int = 0

for tag_key_b in redis_client.scan_iter(match=b'poststag:*'):
    tag_key: str = tag_key_b.decode()
    tag_name: str = tag_key.split(':', 1)[1]

    tags_procesados += 1
    tagsusers_key: str = f'tagsusers:{tag_name}'

    for post_id_b in redis_client.sscan_iter(tag_key_b):
        owner_uid_b = redis_client.hget(post_id_b, 'OwnerUserId')
        if not owner_uid_b:
            continue

        user_id: str = owner_uid_b.decode()
        # Añadir al conjunto tagsusers:<tag>
        redis_client.sadd(tagsusers_key, user_id)
        # Añadir al Bloom filter como "user_id:tag_name"
        bf_user_tag.add(f'{user_id}:{tag_name}')
        pares_insertados += 1

# Copiar Bloom filter a Redis en una sola operación
bf_user_tag.to_redis(redis_client, BF_USER_TAG_KEY)

print(f"Tags procesados: {tags_procesados}")
print(f"Pares user:tag insertados en Bloom filter: {pares_insertados}")
print(f"Tamaño del Bloom filter: {BF_USER_TAG_SIZE} bits")
````

Ahora usamos el Bloom filter para responder rápidamente:

**¿El usuario con id `X` ha hecho una pregunta con el tag `Y`?**

1. Consultamos el Bloom filter con la clave `"X:Y"`.
2. Si el Bloom filter dice **no** → la respuesta es **definitivamente no** (sin más consultas a Redis).
3. Si el Bloom filter dice **posiblemente sí** → confirmamos con `SISMEMBER tagsusers:Y X`.

````python
# Cargar Bloom filter desde Redis una sola vez y reutilizarlo
bf_user_tag: RedisBloomFilter = RedisBloomFilter.from_redis(
    redis_client,
    BF_USER_TAG_KEY,
    size=BF_USER_TAG_SIZE,
    num_hashes=BF_USER_TAG_HASHES,
)


def usuario_tiene_post_con_tag(user_id: str, tag_name: str, verbose: bool = True) -> tuple[bool, bool]:
    """Comprueba si el usuario user_id ha hecho una pregunta con el tag tag_name.
    Devuelve (resultado, se_consulto_redis).
    """
    # Consultar el Bloom filter con "user_id:tag_name"
    if not bf_user_tag.exists(f'{user_id}:{tag_name}'):
        if verbose:
            print(f"  user={user_id}, tag={tag_name}: NO (descartado por Bloom filter)")
        return (False, False)

    # Posible positivo: confirmar con SISMEMBER
    resultado: bool = redis_client.sismember(f'tagsusers:{tag_name}', user_id)
    if verbose:
        print(f"  user={user_id}, tag={tag_name}: {'SÍ' if resultado else 'NO (falso positivo del Bloom filter)'} (confirmado con SISMEMBER)")
    return (resultado, True)


# Ejemplos de consulta: buscar pares válidos (usuario, tag) desde los conjuntos poststag
pares_ejemplo: list[tuple[str, str]] = []

for tag_key_b in redis_client.scan_iter(match=b'poststag:*'):
    tag_key = tag_key_b.decode()
    if redis_client.type(tag_key_b) != b'set' or redis_client.scard(tag_key_b) == 0:
        continue

    tag_name = tag_key.split(':', 1)[1]
    post_ejemplo_b = next(redis_client.sscan_iter(tag_key_b), None)
    if not post_ejemplo_b:
        continue

    owner_user_id_b = redis_client.hget(post_ejemplo_b, 'OwnerUserId')
    if owner_user_id_b:
        pares_ejemplo.append((owner_user_id_b.decode(), tag_name))

    if len(pares_ejemplo) >= 3:
        break

# Añadir pares improbables para ver descartes del Bloom filter
pares_ejemplo.append(('999999999', 'tag-no-existe-demo'))
if pares_ejemplo:
    pares_ejemplo.append((pares_ejemplo[0][0], 'tag-inexistente-xyz'))

print("Consultas de ejemplo:")
for uid, tag in pares_ejemplo:
    usuario_tiene_post_con_tag(uid, tag, verbose=True)
````

### Uso como pre-filtro antes de `SISMEMBER`

En lugar de consultar siempre `SISMEMBER tagsusers:<tag> <user_id>`, usamos el Bloom filter como pre-filtro:

1. Consultamos el Bloom filter con la clave `"user_id:tag_name"`.
2. Si dice que **no existe** → la respuesta es definitivamente no. Nos ahorramos la consulta a Redis.
3. Si dice que **posiblemente existe** → ejecutamos `SISMEMBER tagsusers:<tag> <user_id>` para confirmar.

El ahorro es significativo cuando la mayoría de consultas son negativas (el par usuario-tag no existe).

````python
# Versión directa (sin Bloom filter): siempre SISMEMBER

def usuario_tiene_post_con_tag_directo(user_id: str, tag_name: str) -> bool:
    """Comprueba directamente con SISMEMBER si el usuario hizo una pregunta con el tag."""
    return redis_client.sismember(f'tagsusers:{tag_name}', user_id)


# Comparación rápida con un par de ejemplo
if pares_ejemplo:
    uid0, tag0 = pares_ejemplo[0]
    r_bf = usuario_tiene_post_con_tag(uid0, tag0, verbose=True)
    r_dir = usuario_tiene_post_con_tag_directo(uid0, tag0)
    print(f"Con Bloom filter: resultado={r_bf[0]}, consultó Redis={r_bf[1]}")
    print(f"Directo (SISMEMBER): resultado={r_dir}")
else:
    print("No hay pares de ejemplo disponibles.")
````

### Comparación de rendimiento: `SISMEMBER` directo vs pre-filtro Bloom

Medimos la diferencia de tiempo en consultas aleatorias de pares (`user_id`, `tag_name`) comparando:

- **Estrategia directa**: `SISMEMBER tagsusers:<tag> <user_id>` siempre
- **Estrategia con Bloom filter**: solo `SISMEMBER` cuando el Bloom filter detecta un posible positivo

````python
import random
import time

# usuarios y tags disponibles (usamos números aproximados para simular)
users_disponibles: int = 70000
tags_disponibles: list[str] = list({k.decode().split(':', 1)[1]
                                    for k in redis_client.scan_iter(match=b'tagsusers:*')})

n_pairs: int = 10000
# Generar pares aleatorios (la mayoría no existirán -> buen caso para Bloom filter)
pares_consulta: list[tuple[str, str]] = [
    (str(random.randint(1, users_disponibles)), random.choice(tags_disponibles))
    for _ in range(n_pairs)
]

print(f"Consultas: {len(pares_consulta)} pares (usuario, tag)")
print(f"Usuarios disponibles: {users_disponibles}")
print(f"Tags disponibles: {len(tags_disponibles)}")

# Estrategia A: SISMEMBER siempre
t0: float = time.time()
positivos_directo: int = 0
for user_id, tag_name in pares_consulta:
    if usuario_tiene_post_con_tag_directo(user_id, tag_name):
        positivos_directo += 1
t_directo: float = time.time() - t0

# Estrategia B: pre-filtro Bloom + SISMEMBER solo cuando procede
t0 = time.time()
sismember_bloom: int = 0
positivos_bloom: int = 0
descartados_bloom: int = 0
for user_id, tag_name in pares_consulta:
    if not bf_user_tag.exists(f'{user_id}:{tag_name}'):
        descartados_bloom += 1
        continue
    # Posible positivo: confirmar con SISMEMBER
    sismember_bloom += 1
    if redis_client.sismember(f'tagsusers:{tag_name}', user_id):
        positivos_bloom += 1
t_bloom: float = time.time() - t0

print(f"\n=== Comparación: SISMEMBER directo vs Bloom + SISMEMBER ===")
print(f"SISMEMBER directo: {len(pares_consulta)} llamadas en {t_directo:.3f}s")
print(f"Bloom + SISMEMBER: {sismember_bloom} llamadas en {t_bloom:.3f}s, ventaja: {t_directo/t_bloom:.2f}x")
print(f"Descartados por Bloom (sin consultar Redis): {descartados_bloom} ({descartados_bloom/len(pares_consulta)*100:.1f}%)")
print(f"Positivos directos: {positivos_directo}")
print(f"Positivos con Bloom: {positivos_bloom}")
````
